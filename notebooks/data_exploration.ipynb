# Patients with cancer

```python
df = train_metadata.loc[train_metadata['target']==1].copy().sort_values('age_approx').reset_index(drop=True)
```

```python
df
```

```python
no_duplicates = df.drop_duplicates(subset='patient_id', keep="last").reset_index(drop=True)
```

```python
cancer_patients = list(no_duplicates['patient_id'].to_numpy())
len(cancer_patients)
```

```python
round(len(no_duplicates)*0.1)
```

```python
sample_val = no_duplicates.sample(n=round(len(no_duplicates)*0.1))
```

```python
sample_val['patient_id'].to_numpy()
```

# Patients with no cancer

```python
df_no = train_metadata[~train_metadata['patient_id'].isin(cancer_patients)].reset_index(drop=True)
```

```python
df_no
```

```python
no_cancer_duplicates = df_no.drop_duplicates(subset='patient_id', keep="last").reset_index(drop=True)
```

```python
no_cancer_duplicates
```

```python
no_cancer_patients = list(no_cancer_duplicates['patient_id'].to_numpy())
```

```python
len(no_cancer_patients)
```

```python
sample_val_cancer = no_cancer_duplicates.sample(n=round(len(no_cancer_duplicates)*0.1))
```

```python
sample_val_cancer
```

# Validation set

```python
val_patients = list(sample_val['patient_id'].to_numpy()) + list(sample_val_cancer['patient_id'].to_numpy())
```

```python
validation_dataset = train_metadata[train_metadata['patient_id'].isin(val_patients)].reset_index(drop=True)
```

```python
validation_dataset.to_csv('val_metadata.csv')
```
